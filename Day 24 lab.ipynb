```python
# LAB 1
import matplotlib.pyplot as plt
import numpy as np

# Data for square footage and selling prices
square_footage = np.array([1200, 1400, 1600, 1800, 2000, 2200, 2400, 2600, 2800, 3000])
selling_prices = np.array([250, 290, 315, 380, 410, 450, 500, 525, 570, 610])

# Plotting the scatter plot
plt.figure(figsize=(10, 5))
plt.scatter(square_footage, selling_prices, color='blue', marker='o')
plt.xlabel('Square Footage')
plt.ylabel('Selling Prices (in $1000)')
plt.title('Relationship between House Size and Selling Prices')
plt.grid(True)
plt.tight_layout()
plt.show()
```

The scatter plot shows a clear positive correlation between the square footage of houses and their selling prices, indicating that larger houses tend to sell for higher prices. The trend appears to be linear, suggesting a consistent increase in price with an increase in house size. This information can be useful for predicting the selling prices of houses based on their square footage in this neighborhood

```python
# lAB 2
import matplotlib.pyplot as plt

# Data for income sources and monthly income
income_sources = ['Salary', 'Freelance', 'Investments', 'Rental', 'Other']
monthly_income = [5000, 1500, 1000, 600, 400]

# Plotting the pie chart
plt.figure(figsize=(8, 8))
plt.pie(monthly_income, labels=income_sources, autopct='%1.1f%%', startangle=140, colors=['blue', 'green', 'red', 'purple', 'orange'])
plt.title('Distribution of Monthly Income by Source')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

The pie chart reveals that the majority of your monthly income, 62.5%, comes from your salary, making it the primary source. Freelance work and investments contribute 18.8% and 12.5%, respectively, while rental income and other sources make up smaller portions. This distribution indicates a diversified income structure, with a significant reliance on salary but additional support from other income streams
