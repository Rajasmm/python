# Generating a Bar Chart with Regions and Sales
Create a bar chart to represent the sales data for different regions and calculate the percentage of total sales each region contributes. Provide your conclusion based on the chart.

**Input:**
```python
regions = ['North', 'South', 'East', 'West', 'Central']
sales = [150000, 230000, 120000, 170000, 200000]

````python
import matplotlib.pyplot as plt

# Data for regions and sales
regions = ['North', 'South', 'East', 'West', 'Central']
sales = [150000, 230000, 120000, 170000, 200000]

# Calculate total sales and percentage contribution
total_sales = sum(sales)
percentage_contribution = [sale / total_sales * 100 for sale in sales]

# Plotting the bar chart
plt.figure(figsize=(10, 5))
plt.bar(regions, sales, color='purple')
plt.xlabel('Regions')
plt.ylabel('Sales ($)')
plt.title('Sales Data for Different Regions')
plt.grid(axis='y')
plt.tight_layout()
plt.show()

# Print percentage contribution
for region, percentage in zip(regions, percentage_contribution):
    print(f"{region}: {percentage:.2f}% of total sales")
````
